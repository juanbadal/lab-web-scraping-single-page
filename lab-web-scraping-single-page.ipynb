# Lab | WebScrapping Single Page

```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import requests
```

```python
url  = 'https://www.billboard.com/charts/hot-100/'

response = requests.get(url)
response.status_code

soup = BeautifulSoup(response.content, "html.parser")
```

```python
songs = [elem.get_text().replace('\n', '').replace('\t', '') for elem in soup.select('div.chart-results-list div.o-chart-results-list-row-container li.lrv-u-width-100p h3')]
songs
```

```python
artists = [elem.get_text().replace('\n', '').replace('\t', '') for elem in soup.select('div.chart-results-list div.o-chart-results-list-row-container li.lrv-u-width-100p span.u-letter-spacing-0021')]
artists
```

```python
songs_df = pd.DataFrame({'rank': range(1, 101), 'song_title': songs, 'artist': artists})
songs_df
```

```python
songs_df.to_csv('billboard100.csv')
```
